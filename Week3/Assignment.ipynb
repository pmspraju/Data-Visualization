```python
#import numpy and pandas
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
#read excel in to a data frame
df_raw = pd.read_csv('C:\Users\c830587\Installed\Data-Visualization\Data\Topic_Survey_Assignment.csv')
df_raw.set_index(df_raw.columns[0], inplace=True)
df_raw.index.name = ''
df_raw.head()
```

```python
#sor the dataframe by Very interested column
sortcol = df_raw.columns[0]
df_sort = df_raw.sort_values(by=sortcol, ascending=False)
df_sort.head()
```

```python
df_tran = df_sort.transpose()
df_tran.head()
```

```python
import matplotlib.pyplot as plt
categories = len(list(df_tran.columns))

# create plot
fig, ax = plt.subplots(1,1,figsize=(20,8))
index = np.arange(categories)
bar_width = 0.26
opacity = 0.8
colormap = ['#5cb85c','#5bc0de','#d9534f']
ind_len = len(list(df_tran.index))
bar = range(ind_len)
for i in range(ind_len):
    bar = plt.bar(index + i*bar_width, list(df_tran.iloc[i]), bar_width,
    #align='center',
    color=colormap[i],
    label= df_tran.index[i])

titl = 'Percentage of Respondents' + "' " + 'Interest in Data Science Areas'
plt.title(titl,fontsize=16)
plt.xticks(index + bar_width, list(df_tran.columns),rotation='vertical',fontsize=14)
plt.legend(fontsize=14)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_yticklabels([])
ax.yaxis.set_ticks_position('none')
 
plt.tight_layout()
plt.show()
```
